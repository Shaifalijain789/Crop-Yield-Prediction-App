```python
weather_cols = [
    'T2M', 'T2M_MIN', 'T2M_MAX',
    'RH2M', 'WS2M', 'PRECTOTCORR',
    'ALLSKY_SFC_SW_DWN'
]

df[weather_cols].hist(figsize=(14,10), bins=30)
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='T2M', y='YIELD_TON_HA', data=df)
plt.title("Avg Temperature vs Yield")
plt.show()
```

```python
sns.scatterplot(x='PRECTOTCORR', y='YIELD_TON_HA', data=df)
plt.title("Rainfall vs Yield")
plt.show()
```

```python
sns.scatterplot(x='ALLSKY_SFC_SW_DWN', y='YIELD_TON_HA', data=df)
plt.title("Solar Radiation vs Yield")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='SEASON', y='YIELD_TON_HA', data=df)
plt.title("Season-wise Yield Comparison")
plt.show()
```

```python
numeric_cols = [
    'AREA_HECTARE', 'YEAR', 'T2M', 'T2M_MIN', 'T2M_MAX',
    'RH2M', 'WS2M', 'PRECTOTCORR',
    'ALLSKY_SFC_SW_DWN', 'YIELD_TON_HA'
]

corr = df[numeric_cols].corr()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
