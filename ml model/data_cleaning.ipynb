```python
for col in weather_numeric_col:
    weather_df[col]=pd.to_numeric(
        weather_df[col],
        errors='coerce'
    )
```

```python
weather_df=weather_df.sort_values("DATE")
```

```python
weather_df.columns
```

```python
weather_df.rename(columns={"CITY":"STATE"},inplace=True)
```

```python
weather_monthly = (
    weather_df
    .groupby(["STATE", "YEAR", "MONTH"])
    .agg({
        "T2M": "mean",
        "T2M_MIN": "mean",
        "T2M_MAX": "mean",
        "RH2M": "mean",
        "WS2M": "mean",
        "PRECTOTCORR": "sum",
        "ALLSKY_SFC_SW_DWN": "mean"
    })
    .reset_index()
)

```

```python
print("\nFinal Daily Weather Shape:", weather_df.shape)
print("\nFinal Yearly monthly Shape:", weather_monthly.shape)
print("\nMissing values (Yearly):")
print(weather_monthly.isnull().sum())
```

```python
crop_rabi=crop_df[crop_df["SEASON"]=="Rabi"]
crop_kharif=crop_df[crop_df["SEASON"]=="Kharif"]
crop_summer=crop_df[crop_df["SEASON"]=="Summer"]
crop_autumn=crop_df[crop_df["SEASON"]=="Autumn"]
```

```python
crop_df["YEAR"].unique()
```

```python
SEASON_MONTHS = {
    "Kharif":  [6, 7, 8, 9],
    "Autumn":  [9, 10, 11],
    "Rabi":    [12, 1, 2, 3],
    "Summer":  [3, 4, 5]
}
```

```python

def assign_crop_year(season, month, year):
    if season == "Rabi" and month in [1, 2, 3]:
        return year - 1
    else:
        return year
```

```python
rabi_weather = weather_df[weather_df["MONTH"].isin(SEASON_MONTHS["Rabi"])].copy()

rabi_weather["SEASON"] = "Rabi"
rabi_weather["cROP_YEAR"] = rabi_weather.apply(
    lambda r: assign_crop_year("Rabi", r["MONTH"], r["YEAR"]),
    axis=1
)
```

```python
rabi_weather.columns
```

```python
kharif_weather = weather_df[weather_df["MONTH"].isin(SEASON_MONTHS["Kharif"])].copy()
kharif_weather["SEASON"] = "Kharif"
```

```python
autumn_weather = weather_df[weather_df["MONTH"].isin(SEASON_MONTHS["Autumn"])].copy()
autumn_weather["SEASON"] = "Autumn"
```

```python
summer_weather = weather_df[weather_df["MONTH"].isin(SEASON_MONTHS["Summer"])].copy()
summer_weather["SEASON"] = "Summer"
```

```python
weather_agg = {
    "T2M": "mean",
    "T2M_MIN": "mean",
    "T2M_MAX": "mean",
    "RH2M": "mean",
    "WS2M": "mean",
    "PRECTOTCORR": "sum",
    "ALLSKY_SFC_SW_DWN": "mean"
}

def aggregate_weather(df):
    return (
        df
        .groupby(["STATE","YEAR","SEASON"])
        .agg(weather_agg)
        .reset_index()
    )

rabi_weather_agg   = aggregate_weather(rabi_weather)
kharif_weather_agg = aggregate_weather(kharif_weather)
autumn_weather_agg = aggregate_weather(autumn_weather)
summer_weather_agg = aggregate_weather(summer_weather)
```

```python
rabi_weather_agg.columns
```

```python
rabi_merged_df= crop_rabi.merge(
    rabi_weather_agg,
    left_on=["STATE","YEAR", "SEASON"],
    right_on=["STATE","YEAR", "SEASON"],
    how="inner"
)

```

```python
kharif_merged_df= crop_kharif.merge(
    kharif_weather_agg,
    left_on=["STATE","YEAR", "SEASON"],
    right_on=["STATE","YEAR", "SEASON"],
    how="inner"
)
```

```python
autumn_merged_df= crop_autumn.merge(
    autumn_weather_agg,
    left_on=["STATE","YEAR", "SEASON"],
    right_on=["STATE","YEAR", "SEASON"],
    how="inner"
)
```

```python
summer_merged_df= crop_summer.merge(
    summer_weather_agg,
    left_on=["STATE","YEAR", "SEASON"],
    right_on=["STATE","YEAR", "SEASON"],
    how="inner"
)
```

```python
final_df = pd.concat(
    [kharif_merged_df, rabi_merged_df, autumn_merged_df, summer_merged_df],
    axis=0,
    ignore_index=True
)

print("Final dataset shape:", final_df.shape)
```
